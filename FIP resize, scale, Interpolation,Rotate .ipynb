```python
scale = 50
W1 = int(i.shape[1]*scale/100)
H1 = int(i.shape[0]*scale/100)
img2 = cv2.resize(i, (W1, H1), fx=0.4, fy=0.5, interpolation = cv2.INTER_AREA)
cv2_imshow(i)
```

```python
center = (W1//2,H1//2)
rotate_matrix = cv.getRotationMatrix2D(center = center, angle=90 , scale =1)
cv.warpAffine (scr=i , M= rotate_matrix , dsize=(W1,H1))
```
