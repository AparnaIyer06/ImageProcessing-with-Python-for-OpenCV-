```python
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/gdrive')
import os
```

```python
os.chdir('/gdrive/My Drive/Dataset')
I = plt.imread('bubbles.tif')
print(I.shape)

plt.imshow(I)
plt.show()

import cv2

I1 = cv2.cvtColor(I,cv2.COLOR_BGR2GRAY)
#we convert 3d image to 2d Image


I1.shape

I2 = I1.ravel()

I2.shape

plt.hist(I1.ravel(),256,[0,256])
plt.show()
```

```python
I.shape
I1.shape
```

```python
histr = cv2.calcHist([I1],[0],None,[256],[0,256])
plt.plot(histr)
plt.show()
```

Bins are like counter

```python
I3= cv2.equalizeHist(I1)
plt.hist(I3.ravel(),256,[0,256])
plt.show()
```

```python
print(I3.shape)
```

```python
plt.imshow(I3, cmap="gray")
```

```python

plt.figure(figsize=[10,10])
plt.subplot(2,2,1)
plt.imshow(I)
plt.title("Original Image")
plt.xlabel('X')
plt.ylabel('Y')
plt.subplot(2,2,2)
plt.hist(I1.ravel(),256,[0,256])
plt.title("Histogram of Original Image")
plt.xlabel('Gray Levels')
plt.ylabel('Frequency of occurence')
plt.subplot(2,2,3)
plt.imshow(I3,cmap='gray')
plt.title("Histogram Equalized Image")
plt.subplot(2,2,4)
plt.hist(I3.ravel(),256,[0,256])
plt.title("Equalized Histogram")
plt.show()
```
